```python
# sum of n numbers
def calculate_sum(num):
    total = 0
    for i in range(1, num + 1):
        total += i   # total = total + i
    return total

def main():
    n = int(input("Enter number: "))
    print("Sum:", calculate_sum(n))

main()
```

```python
#Finding the even numbers from 1 to n
def even_nos(n):
    for i in range(1,n+1):
        if i%2==0:
            print(i)
def main():
    n= int(input("Enter numbers:"))
    print(even_nos(n))
main()
```

```python
contact = {}
def add_contact(name: str, phone: str):
    if name not in contact:
        contact[name] = phone
        print("Contact added successfully")
    else:
        print("Contact already exists")

def update_contact(name: str, phone: str):
    if name in contact:
        contact[name] = phone
        print("Contact updated successfully")
    else:
        print("Contact does not exist")

def get_contact(name: str):
    if name in contact:
        return contact[name]
    else:
        print("Contact does not exist")
        return None

def display_contacts():
    if contact:
        for name, phone in contact.items():
            print(name, ":", phone)
    else:
        print("No contacts available")
add_contact("Nandini","8187846200")
add_contact("Pavani","8187846200")
add_contact("ravi","7989648969")
add_contact("Lydia","7396646178")
update_contact("Pavani","9484924849")
print(get_contact("ravi"))
display_contacts()
```
