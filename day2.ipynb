# Data types

* float -> stores the numeric values by using desimals (12.3, 3.5)
* string -> combination of characters
* boolean -> yes/no, true/false, 0/1
* char -> 'a','b','c',.....'z'

# Loops : 
--> When ever we need to do the task multiple times, we need to use the loops
* for : When ever we know the no of iterations.
* while : when ever we dont know the no of iterations.
 # Example: 
 * Cooking food for family(for) bcz we know the count of people.

# if & else 
* these are used to make desitions in a program.
* It executes diffrtrnt codes based on the conditions

```python
# to find max of 2 num
def find_max(a,b):
    if a > b:
        print("Maximum :",a)
    else:
        print("Maximum : ",b)
def main():
    a= int(input("Enter a num: "))
    b= int(input("Enter b number: " ))
    find_max(a,b)
main()
```

```python
def student_grade(marks):
    if marks>=90:
        print("Grade 0")
    elif marks >=80:
        print("Grade A+")
    elif marks >= 70:
        print("Grade A")
    elif marks >= 60:
        print("Grade B")
    else:
        print("Fail")
def main():
    k = int(input())
    student_grade(k)
main()
```

```python
# student profile
student_id = " "
name = " "
age = " "
course = " "
while True:
    print("-----Student Profile Menu-----")
    print("1. Add Student Profile")
    print("2. View Student Profile")
    print("3. Update Student Profile")
    print("4. Delete Student Profile")
    print("5. Exit")
    choice = input("Enter your choice (1-5): ")
    if choice == '1':
        student_id = input("Enter Student ID: ")
        name = input("Enter Name: ")
        age = input("Enter Age: ")
        course = input("Enter Course: ")
        print("Student profile added successfully!")
    elif choice == '2':
        if student_id:
            print("Student ID:", student_id)
            print("Name:", name)
            print("Age:", age)
            print("Course:", course)
        else:
            print("No student profile found.")
    elif choice == '3':
        if student_id:
            name = input("Enter Name: ")
            age = input("Enter Age: ")
            course = input("Enter Course: ")
            print("Student profile updated successfully!")
        else:
            print("No student profile found.")
    elif choice == '4':
        student_id = " "
        name = " "
        age = " "
        course = " "
        print("Student profile deleted successfully!")
    elif choice == '5':
        print("Exiting the program. Goodbye!")
        break   
    else:
        print("Invalid choice. Please try again.")
        



```
