```python
# function to check prime num
def is_prime(num):
    if num <=1:
        return false
    for i in range(2,num):
        if num %i==0:
            return False
    return True
def main():
    num = int(input())
    if is_prime(num):
        print("Prime")
    else:
        print("Not prime")
main()
```
