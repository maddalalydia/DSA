```python
# Insertion sort
def insertion_sort(arr):
    for i in range(1,len(arr)):
        key = arr[i]
        j = i - 1
        #Shifting elements
        while j >=0 and arr[j]>key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr
arr = [9,4,2,7]
print(insertion_sort(arr))
```

```python
def insert_into_sorted(arr, x):
    arr.append(0)
    i = len(arr) - 2
    while i >= 0 and arr[i] > x:
        arr[i + 1] = arr[i]
        i -= 1
    arr[i + 1] = x
    return arr
arr = [1, 3, 5, 7]
x = 4
print(insert_into_sorted(arr, x))
```
